真實數據分析

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
print(iris.DESCR)
```

```python
X = iris.data
```

```python
Y = iris.target
```

```python
X[0]
```

```python
X[0][2:]
```

```python
X = X[:, 2:]
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y,
                                                   test_size = 0.2,
                                                   random_state=87) #random_state是為了要讓每次的都不一樣
```

```python
plt.scatter(x_train[:,0], x_train[:,1], c=y_train)
```

```python
from sklearn.svm import SVC
```

```python
clf = SVC()
```

```python
clf.fit(x_train, y_train)
```

```python
y_predict = clf.predict(x_test)
```

```python
plt.scatter(x_test[:,0], x_test[:,1], c=y_predict)
```

```python
plt.scatter(x_test[:,0], x_test[:,1], c=y_predict - y_test) #-y_test 檢查預測結果，正確一個顏色，錯誤另一個顏色
```

## K.means

```python
clf.labels_
```
