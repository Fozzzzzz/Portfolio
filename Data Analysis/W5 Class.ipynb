```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df.head()
```

```python
df.describe()
```

```python
df.corr()
```

```python
df.國文.corr(df.數學)
```

```python
df["總級分"] = df[["國文", "英文", "數學", "自然", "社會"]].sum(1)
```

```python
df.head()
```

sum(1) 代表一行一行加 

sum(0) 代表一列一列加

```python
df["主科"] = df.數學*1.5 + df.英文
```

```python
df.head()
```

```python
df.sort_values(by="總級分", ascending=False).head(20)
```

ascending=False 代表從小排到大

```python
df.sort_values(by=["主科", "總級分"], ascending=False).head(20)
```

by=[1, 2] 先比1在比2

## 開始使用 pandas-datareader 套件

```python
import pandas_datareader.data as web
```

```python
df = web.DataReader("AAPL", "yahoo", start="2012-9-1", end="2017-8-31")
```

```python
df.head()
```

```python
P = df["Adj Close"]
```

```python
P.head()
```

```python
P.plot()
```

## 計算報酬率

```python
P.diff()
```

```python
r = P.diff()/P
```

```python
r.plot()
```

```python
r[-100:].plot()
```

## 移動平均
