**RK4**

$k_1 = f(t_n,y_n)h$

$k_2 = f(t_{n+1/2},y_n+k_1/2)h$

$k_3 = f(t_{n+1/2},y_n+k_2/2)h$

$k_4 = f(t_{n+1},y_n+k_3)h$

$y_{n+1}| = y_n+ (k_1+2k_2+2k_3 +k_4)/6$

```python
def ode_RK4(f,t0,tf,y0,n):
  t = linspace(t0,tf,n)
  y = [y0]
  for i in range(n-1):
    h = t[i+1]-t[i]
    k1 = f(t[i],y[-1])*h
    k2 = f(t[i]+h/2,y[-1]+k1/2)*h
    k3 = f(t[i]+h/2,y[-1]+k2/2)*h
    k4 = f(t[i+1],y[-1]+k3)*h
    y.append(y[-1]+(k1+2*k2+2*k3+k4)/6)
  return np.array(y)
```

**RK2**

$k1 = f(t_n,y_n)h$

$k2 = f(t_{n+1/2}, y_n+k_1/2)h$

$y_{n+1} = y_n + k_2$

```python
def ode_RK2(f,t0,tf,y0=0,n=100):
  t = linspace(t0,tf,n)
  y = list()
  y.append(y0)
  for i in range(n-1):
    h = t[i+1]-t[i]
    k1 = f(t[i],y[i])*h
    k2 = f(t[i]+h/2,y[i]+k1/2)*h
    y.append(y[-1]+k2)
  return np.array(y)
```

$y'(t) = f(t,y)$

$y(0) = y_0$

```python
def ode_Euler(f,t0,tf,y0=0,n=100):
  t = linspace(t0,tf,n)
  y = list()
  y.append(y0)
  for i in range(n-1):
    h = t[i+1]-t[i]
    #y_n+1 = y_n + h*f(t_n,y_n)
    y.append(y[-1]+h*f(t[i],y[-1]))
  return np.array(y)
```

```python
def f(t,y):
  return 4-3*y
```

```python
def F(t):
  return -1.0/3*exp(-3*t)+4.0/3
```

```python
y_euler = ode_Euler(f,0,2,1,400)
y_RK2 = ode_RK2(f,0,2,1,400)
y_RK4 = ode_RK4(f,0,2,1,400)
```

```python
t = np.linspace(0,2,400)
yt = F(t)
```

```python
plot(t,yt-y_euler,t,yt-y_RK2,'r',t,yt-y_RK4,'g')
```
